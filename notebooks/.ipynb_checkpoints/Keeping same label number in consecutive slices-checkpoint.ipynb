# Keeping same label number in consecutive slices

This notebook shows how to segment objects slice-by-slice and then assign the same labels to overlapping objects to create a 3D-segmentation. 

For demonstration purposes I use three consecutive slices of the dataset by [Ozpolat, B. et al](https://doi.org/10.5281/zenodo.1063531) licensed [CC BY 4.0](https://creativecommons.org/licenses/by/4.0/legalcode)
(single timepoint, single channel, rescaled)

```python
from stardist.matching import group_matching_labels
import napari
from skimage.io import imread
import os
from tifffile import imwrite
```

```python
labels_ungrouped = imread("imgs/stardist_ungrouped_labels.tif")
```

```python
viewer = napari.Viewer()
```

```python
viewer.add_labels(labels_ungrouped, name = 'labels_ungrouped')
```

```python
napari.utils.nbscreenshot(viewer)
```

_(I created those labels using one of the pretrained 2D model of napari-stardist)_

Now we have objects where the label number in consecutive slices is not similar. We can use the [`group_matching_labels`](https://github.com/stardist/stardist/blob/master/stardist/matching.py) function from stardist to change that.

```python
# match labels in consecutive frames (-> simple IoU track)
labels_grouped = group_matching_labels(labels_ungrouped)
```

```python
viewer.add_labels(labels_grouped, name = 'labels_grouped')
```

```python
napari.utils.nbscreenshot(viewer)
```

You can also adjust the __threshold__ in the `group_matching_labels` function to your needs. It is set by default very low to `thresh=1e-10`. But you could modify it for example to 0.3 meaning the labels are matched when 30% overlap like this:

```python
# match labels in consecutive frames (-> simple IoU track)
labels_grouped_modified_tresh = group_matching_labels(labels_ungrouped, thresh = 0.3)
```

```python
viewer.add_labels(labels_grouped_modified_tresh, name = 'labels_grouped_modified_tresh')
```

the result looks like this:

```python
napari.utils.nbscreenshot(viewer)
```

Saving the result to a folder:

```python
grouped_labels_folder = "data/2_annotations/name/"
filename = "grouped_labels"
```

```python
save_dir =  "data/2_annotations/name/"
prefix = "grouped_labels"
```

```python
imwrite(os.path.join(save_dir, prefix  + '.tif'), labels_grouped_modified_tresh)
```
